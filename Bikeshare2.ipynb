```python
import numpy as np
import cv2
import matplotlib.pylab as plt
import pandas as pd
from datetime import datetime

from google.colab import drive
drive.mount('/content/drive/')

import os
# Change working directory to the folder where your CSV lives
os.chdir('/content/drive/MyDrive/Colab Notebooks/AI_in_BE_Python_Labs')

# (Optional alternative)
# %cd /content/drive/My Drive/Colab Notebooks/AI_in_BE_Python_Labs/

# -------------------------------------------------------------------
# Data source: Google Drive
# Path: My Drive > Colab Notebooks > AI_in_BE_Python_Labs
# File: 202510-capitalbikeshare-tripdata.csv
# -------------------------------------------------------------------
file_location = '202510-capitalbikeshare-tripdata.csv'

# Keep your original variable name for compatibility with prior labs
imdb_processed = pd.read_csv(file_location)

# Create a clearer name for this specific project
bikeshare_df = imdb_processed.copy()

# Display the first few rows to confirm successful load
bikeshare_df.head()

# -----------------------------------------------------------
# 1. BASIC DATA INSPECTION
# -----------------------------------------------------------

# Overview of columns, types, and non-null counts
print("=== DataFrame info ===")
bikeshare_df.info()

print("\n=== Sample rows ===")
display(bikeshare_df.head())

print("\n=== Summary statistics (numeric) ===")
display(bikeshare_df.describe())

print("\n=== Summary statistics (including object columns) ===")
display(bikeshare_df.describe(include='all'))

print("\n=== Missing values per column ===")
print(bikeshare_df.isna().sum())

# -----------------------------------------------------------
# 2. DATA CLEANING & FEATURE ENGINEERING
# -----------------------------------------------------------

# Convert started_at and ended_at to datetime
bikeshare_df['started_at'] = pd.to_datetime(bikeshare_df['started_at'])
bikeshare_df['ended_at']   = pd.to_datetime(bikeshare_df['ended_at'])

# Create trip duration in minutes
bikeshare_df['trip_duration_min'] = (
    (bikeshare_df['ended_at'] - bikeshare_df['started_at'])
    .dt.total_seconds() / 60
)

# Remove trips with non-positive or extreme durations (optional sanity filter)
bikeshare_df = bikeshare_df[
    (bikeshare_df['trip_duration_min'] > 0) &
    (bikeshare_df['trip_duration_min'] <= 180)  # keep trips <= 3 hours
].copy()

# Extract time-based features
bikeshare_df['start_hour']    = bikeshare_df['started_at'].dt.hour
bikeshare_df['start_day']     = bikeshare_df['started_at'].dt.day_name()
bikeshare_df['start_month']   = bikeshare_df['started_at'].dt.month
bikeshare_df['start_date']    = bikeshare_df['started_at'].dt.date
bikeshare_df['is_weekend']    = bikeshare_df['start_day'].isin(['Saturday', 'Sunday'])

# Verify new columns
bikeshare_df[['started_at','ended_at','trip_duration_min','start_hour','start_day','is_weekend']].head()

# -----------------------------------------------------------
# 3. DESCRIPTIVE STATISTICS BY USER TYPE
# -----------------------------------------------------------

print("=== Trip counts by user type ===")
print(bikeshare_df['member_casual'].value_counts())

print("\n=== Average trip duration (minutes) by user type ===")
print(bikeshare_df.groupby('member_casual')['trip_duration_min'].mean().round(2))

print("\n=== Trip counts by bike type ===")
print(bikeshare_df['rideable_type'].value_counts())

print("\n=== Average trip duration by bike type and user type ===")
display(
    bikeshare_df
    .groupby(['rideable_type', 'member_casual'])['trip_duration_min']
    .mean()
    .round(2)
    .unstack()
)

# -----------------------------------------------------------
# 4. TEMPORAL PATTERNS (BY HOUR & DAY OF WEEK)
# -----------------------------------------------------------

# Order for plotting days of week
day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
bikeshare_df['start_day'] = pd.Categorical(bikeshare_df['start_day'], categories=day_order, ordered=True)

# Trips by hour and user type
hourly_counts = (
    bikeshare_df
    .groupby(['start_hour','member_casual'], observed=False)['ride_id']
    .count()
    .reset_index()
    .rename(columns={'ride_id':'trip_count'})
)

# Trips by day of week and user type
daily_counts = (
    bikeshare_df
    .groupby(['start_day','member_casual'], observed=False)['ride_id']
    .count()
    .reset_index()
    .rename(columns={'ride_id':'trip_count'})
)

hourly_counts.head(), daily_counts.head()

# -----------------------------------------------------------
# 5. VISUALIZATION: TRIP DURATION
# -----------------------------------------------------------

plt.figure(figsize=(8,5))
plt.hist(bikeshare_df['trip_duration_min'], bins=50)
plt.xlabel('Trip duration (minutes)')
plt.ylabel('Number of trips')
plt.title('Distribution of Trip Durations (All Users)')
plt.tight_layout()
plt.show()

# Log-scale view (optional, can help see shorter vs longer trips)
plt.figure(figsize=(8,5))
plt.hist(bikeshare_df['trip_duration_min'], bins=50, log=True)
plt.xlabel('Trip duration (minutes)')
plt.ylabel('Number of trips (log scale)')
plt.title('Trip Duration Distribution (Log Scale)')
plt.tight_layout()
plt.show()

# -----------------------------------------------------------
# 6. VISUALIZATION: USER TYPE & TIME-OF-DAY PATTERNS
# -----------------------------------------------------------

# Bar chart of total trips by user type
user_counts = bikeshare_df['member_casual'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(user_counts.index, user_counts.values)
plt.xlabel('User type')
plt.ylabel('Number of trips')
plt.title('Trips by User Type')
plt.tight_layout()
plt.show()

# Line plot of trips by hour for each user type
plt.figure(figsize=(10,6))
for utype in hourly_counts['member_casual'].unique():
    subset = hourly_counts[hourly_counts['member_casual'] == utype]
    plt.plot(subset['start_hour'], subset['trip_count'], marker='o', label=utype)

plt.xlabel('Hour of day (023)')
plt.ylabel('Number of trips')
plt.title('Trips by Hour of Day and User Type')
plt.legend(title='User type')
plt.xticks(range(0,24))
plt.tight_layout()
plt.show()

# Bar chart of trips by day of week and user type
plt.figure(figsize=(10,6))

# For stacked or side-by-side bars, pivot the table
daily_pivot = daily_counts.pivot(index='start_day', columns='member_casual', values='trip_count').fillna(0)

# Side-by-side bars
x = np.arange(len(daily_pivot.index))
width = 0.35

plt.bar(x - width/2, daily_pivot['member'],  width, label='member')
plt.bar(x + width/2, daily_pivot['casual'],  width, label='casual')

plt.xticks(x, daily_pivot.index, rotation=45)
plt.xlabel('Day of week')
plt.ylabel('Number of trips')
plt.title('Trips by Day of Week and User Type')
plt.legend()
plt.tight_layout()
plt.show()

# -----------------------------------------------------------
# 7. STATION-LEVEL INSIGHTS (OPTIONAL)
# -----------------------------------------------------------

# Top 10 start stations by number of trips
top_start = (
    bikeshare_df['start_station_name']
    .value_counts()
    .head(10)
)

print("=== Top 10 start stations ===")
print(top_start)

plt.figure(figsize=(10,6))
plt.barh(top_start.index[::-1], top_start.values[::-1])
plt.xlabel('Number of trips')
plt.title('Top 10 Start Stations')
plt.tight_layout()
plt.show()

# Top 10 end stations by number of trips
top_end = (
    bikeshare_df['end_station_name']
    .value_counts()
    .head(10)
)

print("\n=== Top 10 end stations ===")
print(top_end)

plt.figure(figsize=(10,6))
plt.barh(top_end.index[::-1], top_end.values[::-1])
plt.xlabel('Number of trips')
plt.title('Top 10 End Stations')
plt.tight_layout()
plt.show()

import numpy as np
import pandas as pd
import matplotlib.pylab as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression, PoissonRegressor, Ridge
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.preprocessing import LabelEncoder

# -------------------------------------------------------------------
# 1. Build stationday dataset (target = same-day trip count)
# -------------------------------------------------------------------

# Ensure date column exists (from previous code: start_date is a datetime.date)
if 'start_date' not in bikeshare_df.columns:
    bikeshare_df['start_date'] = bikeshare_df['started_at'].dt.date

# Aggregate to stationday-level (originating trips per station per day)
station_day_df = (
    bikeshare_df
    .groupby(['start_station_id', 'start_station_name', 'start_date'])
    .agg(
        trip_count=('ride_id', 'count'),
        # optional: mean trip duration per station-day
        avg_trip_duration_min=('trip_duration_min', 'mean')
    )
    .reset_index()
)

station_day_df['start_date'] = pd.to_datetime(station_day_df['start_date'])

# -------------------------------------------------------------------
# 2. Feature engineering (calendar + historical usage + station context)
# -------------------------------------------------------------------

# Calendar-based features
station_day_df['day_of_week'] = station_day_df['start_date'].dt.dayofweek  # Monday=0,...,Sunday=6
station_day_df['month'] = station_day_df['start_date'].dt.month
station_day_df['is_weekend'] = station_day_df['day_of_week'].isin([5, 6]).astype(int)

# Station-level context: global mean trips per station
station_means = station_day_df.groupby('start_station_id')['trip_count'].mean().rename('station_global_mean')
station_day_df = station_day_df.merge(station_means, on='start_station_id', how='left')

# Sort for proper lag / rolling computation
station_day_df = station_day_df.sort_values(['start_station_id', 'start_date']).reset_index(drop=True)

# Historical usage features per station
def add_lag_features(df, group_col, target_col='trip_count'):
    df = df.copy()
    # Lag features (previous day, previous week)
    df['lag_1'] = df.groupby(group_col)[target_col].shift(1)
    df['lag_7'] = df.groupby(group_col)[target_col].shift(7)

    # Rolling windows (past 7 days, excluding today)
    df['rolling_7_mean'] = (
        df.groupby(group_col)[target_col]
        .shift(1)
        .rolling(window=7, min_periods=3)
        .mean()
    )
    df['rolling_7_std'] = (
        df.groupby(group_col)[target_col]
        .shift(1)
        .rolling(window=7, min_periods=3)
        .std()
    )

    return df

station_day_df = add_lag_features(station_day_df, group_col='start_station_id', target_col='trip_count')

# -------------------------------------------------------------------
# 3. Define next-day prediction target (supervised learning)
# -------------------------------------------------------------------

# Target: next-day trip count for the same station
station_day_df['y_next_day'] = (
    station_day_df
    .groupby('start_station_id')['trip_count']
    .shift(-1)
)

# Drop rows where we don't have target or key lag features
model_df = station_day_df.dropna(
    subset=['y_next_day', 'lag_1', 'lag_7', 'rolling_7_mean']
).copy()

# -------------------------------------------------------------------
# 4. Encode station and day-of-week (for models that need numeric inputs)
# -------------------------------------------------------------------

# Label-encode station_id
station_le = LabelEncoder()
model_df.loc[:, 'station_id_encoded'] = station_le.fit_transform(
    model_df['start_station_id'].astype(str)
)


# One-hot encode day_of_week
day_dummies = pd.get_dummies(model_df['day_of_week'].astype(int), prefix='dow', drop_first=True)
model_df = pd.concat([model_df.reset_index(drop=True), day_dummies.reset_index(drop=True)], axis=1)

# -------------------------------------------------------------------
# 5. Train / validation / test splits (time-based)
# -------------------------------------------------------------------

# Use quantiles of date for splitting into early / middle / late periods
dates_sorted = model_df['start_date'].sort_values().unique()
q1 = np.quantile(np.arange(len(dates_sorted)), 0.6)
q2 = np.quantile(np.arange(len(dates_sorted)), 0.8)

cutoff_train = dates_sorted[int(q1)]
cutoff_val = dates_sorted[int(q2)]

train_mask = model_df['start_date'] <= cutoff_train
val_mask = (model_df['start_date'] > cutoff_train) & (model_df['start_date'] <= cutoff_val)
test_mask = model_df['start_date'] > cutoff_val

train_df = model_df[train_mask].reset_index(drop=True)
val_df = model_df[val_mask].reset_index(drop=True)
test_df = model_df[test_mask].reset_index(drop=True)

print("Train size:", train_df.shape)
print("Val size:", val_df.shape)
print("Test size:", test_df.shape)

# -------------------------------------------------------------------
# 6. Define helper functions: metrics + evaluation
# -------------------------------------------------------------------

def wape(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    denom = np.sum(np.abs(y_true))
    if denom == 0:
        return np.nan
    return np.sum(np.abs(y_true - y_pred)) / denom

def evaluate_predictions(y_true, y_pred, name="Model"):
    mae = mean_absolute_error(y_true, y_pred)
    # Removed 'squared=False' as it's deprecated/removed, and compute RMSE manually
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    wape_val = wape(y_true, y_pred)
    print(f"=== {name} ===")
    print(f"MAE:  {mae:.3f}")
    print(f"RMSE: {rmse:.3f}")
    print(f"WAPE: {wape_val:.3f}")
    print()
    return {'MAE': mae, 'RMSE': rmse, 'WAPE': wape_val}

    # -------------------------------------------------------------------
# 7. Baseline 1: Station mean by weekday
# -------------------------------------------------------------------

# Compute station-weekday means on TRAIN ONLY
train_station_weekday_mean = (
    train_df
    .groupby(['start_station_id', 'day_of_week'])['trip_count']
    .mean()
    .rename('station_weekday_mean')
    .reset_index()
)

# Station global means (for fallback)
train_station_global_mean = (
    train_df
    .groupby('start_station_id')['trip_count']
    .mean()
    .rename('station_global_mean_train')
    .reset_index()
)

# Dataset-level global mean
global_mean_train = train_df['trip_count'].mean()

def baseline_station_weekday_mean(df):
    df_ = df.merge(train_station_weekday_mean,
                   on=['start_station_id', 'day_of_week'],
                   how='left')
    df_ = df_.merge(train_station_global_mean,
                    on='start_station_id',
                    how='left')

    # Fallbacks
    # 1) if station-weekday mean is missing, use station global mean
    df_['station_weekday_mean'] = df_['station_weekday_mean'].fillna(df_['station_global_mean_train'])
    # 2) if station global mean is also missing, use overall global mean
    df_['station_weekday_mean'] = df_['station_weekday_mean'].fillna(global_mean_train)

    return df_['station_weekday_mean'].values

# Predictions using baseline 1 (predict next-day count using station-weekday mean of... day t? day t+1?)
# Simple option: use day_of_week(t+1) by incrementing day_of_week by 1 mod 7.
def add_tomorrow_dow(df):
    df = df.copy()
    df['tomorrow_dow'] = (df['day_of_week'].astype(int) + 1) % 7
    return df

test_df_tw = add_tomorrow_dow(test_df)
val_df_tw = add_tomorrow_dow(val_df)

# For the baseline, we will treat "tomorrow_dow" as the day_of_week for lookup.
test_df_for_baseline = test_df_tw.copy()
test_df_for_baseline['day_of_week'] = test_df_for_baseline['tomorrow_dow']

val_df_for_baseline = val_df_tw.copy()
val_df_for_baseline['day_of_week'] = val_df_for_baseline['tomorrow_dow']

y_test_true = test_df['y_next_day'].values
y_val_true = val_df['y_next_day'].values

y_test_pred_baseline1 = baseline_station_weekday_mean(test_df_for_baseline)
y_val_pred_baseline1 = baseline_station_weekday_mean(val_df_for_baseline)

metrics_baseline1_val = evaluate_predictions(y_val_true, y_val_pred_baseline1, name="Baseline 1 (Station + Weekday Mean) - VAL")
metrics_baseline1_test = evaluate_predictions(y_test_true, y_test_pred_baseline1, name="Baseline 1 (Station + Weekday Mean) - TEST")

# -------------------------------------------------------------------
# 8. Baseline 2: Lag-7 (same weekday last week) as prediction
# -------------------------------------------------------------------

# For next-day prediction, we can use lag_7 relative to day t+1,
# which corresponds to trip_count(t+1 - 7) = trip_count(t-6).
# However, to keep it simple (and still meaningful): we will use lag_7 from day t
# as a proxy baseline for y_next_day.

# We need rows where lag_7 is not null (already enforced by model_df construction)
y_test_pred_baseline2 = test_df['lag_7'].values
y_val_pred_baseline2 = val_df['lag_7'].values

metrics_baseline2_val = evaluate_predictions(y_val_true, y_val_pred_baseline2, name="Baseline 2 (Lag-7) - VAL")
metrics_baseline2_test = evaluate_predictions(y_test_true, y_test_pred_baseline2, name="Baseline 2 (Lag-7) - TEST")

# -------------------------------------------------------------------
# 9. Simple regression baseline (linear / Poisson)
# -------------------------------------------------------------------

# Feature set for regression models (no weather; all derived from usage + calendar)
feature_cols = [
    'lag_1',
    'lag_7',
    'rolling_7_mean',
    'station_global_mean',
    'is_weekend',
    'month',
    'station_id_encoded'
] + [c for c in model_df.columns if c.startswith('dow_')]

X_train = train_df[feature_cols].values
X_val = val_df[feature_cols].values
X_test = test_df[feature_cols].values

y_train = train_df['y_next_day'].values
y_val = val_df['y_next_day'].values
y_test = test_df['y_next_day'].values

# Linear Regression baseline
linreg = LinearRegression()
linreg.fit(X_train, y_train)

y_val_pred_lin = linreg.predict(X_val)
y_test_pred_lin = linreg.predict(X_test)

metrics_lin_val = evaluate_predictions(y_val, y_val_pred_lin, name="Linear Regression - VAL")
metrics_lin_test = evaluate_predictions(y_test, y_test_pred_lin, name="Linear Regression - TEST")

# Poisson Regression (for counts; may take longer, and requires non-negative targets)
try:
    poisson = PoissonRegressor(alpha=1.0, max_iter=1000)
    poisson.fit(X_train, y_train)
    y_val_pred_pois = poisson.predict(X_val)
    y_test_pred_pois = poisson.predict(X_test)

    metrics_pois_val = evaluate_predictions(y_val, y_val_pred_pois, name="Poisson Regression - VAL")
    metrics_pois_test = evaluate_predictions(y_test, y_test_pred_pois, name="Poisson Regression - TEST")
except Exception as e:
    print("Poisson Regression failed:", e)
    y_val_pred_pois = None
    y_test_pred_pois = None

# -------------------------------------------------------------------
# 10. Main approach: Gradient-Boosted Decision Trees (HistGradientBoostingRegressor)
# -------------------------------------------------------------------

gbr = HistGradientBoostingRegressor(
    max_depth=8,
    learning_rate=0.05,
    max_iter=500,
    random_state=42
)

gbr.fit(X_train, y_train)

y_val_pred_gbr = gbr.predict(X_val)
y_test_pred_gbr = gbr.predict(X_test)

metrics_gbr_val = evaluate_predictions(y_val, y_val_pred_gbr, name="Gradient Boosted Trees - VAL")
metrics_gbr_test = evaluate_predictions(y_test, y_test_pred_gbr, name="Gradient Boosted Trees - TEST")

# -------------------------------------------------------------------
# 11. Results summary table (for report "Results & Analysis" section)
# -------------------------------------------------------------------

results_summary = pd.DataFrame([
    ['Baseline 1: Station+Weekday Mean (VAL)', metrics_baseline1_val['MAE'], metrics_baseline1_val['RMSE'], metrics_baseline1_val['WAPE']],
    ['Baseline 1: Station+Weekday Mean (TEST)', metrics_baseline1_test['MAE'], metrics_baseline1_test['RMSE'], metrics_baseline1_test['WAPE']],
    ['Baseline 2: Lag-7 (VAL)', metrics_baseline2_val['MAE'], metrics_baseline2_val['RMSE'], metrics_baseline2_val['WAPE']],
    ['Baseline 2: Lag-7 (TEST)', metrics_baseline2_test['MAE'], metrics_baseline2_test['RMSE'], metrics_baseline2_test['WAPE']],
    ['Linear Regression (VAL)', metrics_lin_val['MAE'], metrics_lin_val['RMSE'], metrics_lin_val['WAPE']],
    ['Linear Regression (TEST)', metrics_lin_test['MAE'], metrics_lin_test['RMSE'], metrics_lin_test['WAPE']],
    ['Gradient Boosted Trees (VAL)', metrics_gbr_val['MAE'], metrics_gbr_val['RMSE'], metrics_gbr_val['WAPE']],
    ['Gradient Boosted Trees (TEST)', metrics_gbr_test['MAE'], metrics_gbr_test['RMSE'], metrics_gbr_test['WAPE']],
])

results_summary.columns = ['Model / Split', 'MAE', 'RMSE', 'WAPE']
print("\n=== Results Summary ===")
display(results_summary)

# -------------------------------------------------------------------
# 12. Diagnostic plots (for Results & Error Analysis sections)
# -------------------------------------------------------------------

# Predicted vs actual for main model on TEST
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_test_pred_gbr, alpha=0.4)
plt.xlabel('Actual next-day trip count')
plt.ylabel('Predicted next-day trip count')
plt.title('Gradient Boosted Trees: Predicted vs Actual (TEST)')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()])
plt.tight_layout()
plt.show()

# Residuals distribution
residuals = y_test - y_test_pred_gbr
plt.figure(figsize=(8,5))
plt.hist(residuals, bins=40)
plt.xlabel('Residual (Actual - Predicted)')
plt.ylabel('Frequency')
plt.title('Residual Distribution (Gradient Boosted Trees, TEST)')
plt.tight_layout()
plt.show()

# Error vs station activity level
test_with_pred = test_df.copy()
test_with_pred['y_pred_gbr'] = y_test_pred_gbr
test_with_pred['abs_error'] = np.abs(test_with_pred['y_next_day'] - test_with_pred['y_pred_gbr'])

station_error = (
    test_with_pred
    .groupby('start_station_id')
    .agg(
        mean_abs_error=('abs_error', 'mean'),
        mean_true=('y_next_day', 'mean'),
        count_days=('y_next_day', 'size')
    )
    .reset_index()
)

plt.figure(figsize=(8,5))
plt.scatter(station_error['mean_true'], station_error['mean_abs_error'])
plt.xlabel('Mean true next-day trips per station (TEST)')
plt.ylabel('Mean absolute error per station')
plt.title('Per-Station Error vs Activity Level')
plt.tight_layout()
plt.show()

# Time series example for a single busy station
if not station_error.empty:
    example_station = station_error.sort_values('mean_true', ascending=False)['start_station_id'].iloc[0]
    station_series = test_with_pred[test_with_pred['start_station_id'] == example_station].sort_values('start_date')

    plt.figure(figsize=(10,5))
    plt.plot(station_series['start_date'], station_series['y_next_day'], marker='o', label='Actual next-day trips')
    plt.plot(station_series['start_date'], station_series['y_pred_gbr'], marker='x', label='Predicted next-day trips')
    plt.xlabel('Date')
    plt.ylabel('Trips')
    plt.title(f'Example Station Time Series (Station ID: {example_station})')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # -------------------------------------------------------------------
# 13. Dataset & descriptive stats for "Dataset" and "Introduction" sections
# -------------------------------------------------------------------

print("=== Dataset Summary (station-day level) ===")
print("Total station-days:", station_day_df.shape[0])
print("Number of unique stations:", station_day_df['start_station_id'].nunique())
print("Date range:", station_day_df['start_date'].min(), "to", station_day_df['start_date'].max())
print("Average trips per station-day:", station_day_df['trip_count'].mean())

# Distribution of station-day trip counts (for dataset description)
plt.figure(figsize=(8,5))
plt.hist(station_day_df['trip_count'], bins=40)
plt.xlabel('Trip count per station-day')
plt.ylabel('Frequency')
plt.title('Distribution of Trip Counts per Station-Day')
plt.tight_layout()
plt.show()

# -------------------------------------------------------------------
# 14. Helper: table of model input/output schema (for "Main Approach" section)
# -------------------------------------------------------------------

input_output_schema = pd.DataFrame({
    'Feature / Target': [
        'start_station_id',
        'start_date',
        'day_of_week',
        'month',
        'is_weekend',
        'station_global_mean',
        'lag_1',
        'lag_7',
        'rolling_7_mean',
        'y_next_day (TARGET)'
    ],
    'Type': [
        'categorical (encoded)',
        'date',
        'integer (0=Mon,...,6=Sun)',
        'integer (1-12)',
        'binary (0/1)',
        'continuous',
        'continuous',
        'continuous',
        'continuous',
        'integer (count)'
    ],
    'Description': [
        'Unique identifier of the bikeshare station',
        'Calendar date for the station-day',
        'Day of week for the station-day',
        'Month for the station-day',
        'Whether the station-day is a weekend',
        'Average daily trips for this station over full history',
        'Trips that departed from this station on the previous day',
        'Trips that departed from this station 7 days before',
        'Mean trips over past 7 days (excluding current day)',
        'Target: predicted trips on the next calendar day for this station'
    ]
})

print("\n=== Input/Output Schema (for Main Approach section) ===")
display(input_output_schema)

# -------------------------------------------------------------------
# 15. (Optional) Template metric formulas (for Evaluation Metric section)
# -------------------------------------------------------------------

# These strings can be copied into your report and turned into LaTeX or markdown.

metric_formulas = {
    'MAE': "MAE = (1/n) *  |y_i - _i|",
    'RMSE': "RMSE = sqrt( (1/n) *  (y_i - _i)^2 )",
    'WAPE': "WAPE = (  |y_i - _i| ) / (  |y_i| )"
}

print("\n=== Metric Formulas (for Evaluation Metric section) ===")
for name, formula in metric_formulas.items():
    print(f"{name}: {formula}")

    # Choose a station and a "today" date that exist in model_df
station_id = '31000'  # <-- put a real start_station_id from your data here
today = pd.to_datetime('2025-10-15')  # <-- put a real date from your data here

# Locate the stationday row for that station on that date
row_today = model_df[
    (model_df['start_station_id'] == station_id) &
    (model_df['start_date'] == today)
]

if row_today.empty:
    print("No data for that station/date combination in model_df.")
else:
    # Extract feature vector in the same order used for training
    X_new = row_today[feature_cols].values  # shape (1, num_features)

    # Use the gradient-boosted model to predict next-day trips
    predicted_tomorrow_trips = gbr.predict(X_new)[0]

    print(f"Predicted trips tomorrow for station {station_id} after {today.date()}: {predicted_tomorrow_trips:.2f}")


    # -------------------------------------------------------------------
# Helper: predict trips for a given station and FUTURE calendar date
# using your trained model, with a baseline fallback.
#
# Place this AFTER all your existing code (so model_df, feature_cols,
# gbr, and baseline_station_weekday_mean already exist).
# -------------------------------------------------------------------

import pandas as pd

def predict_trips_for_station_and_date(
    station_id,
    target_date_str,
    model_df,
    feature_cols,
    model,
    use_baseline=True
):
    """
    Predict next-day trips for a given station on a specific future calendar date.

    How it works:
      - Your model was trained so that each row in model_df (a station-day)
        predicts the NEXT day (y_next_day) for that station.
      - To predict trips on target_date T, we:
          1) Look for the row with date T-1 for that station in model_df.
          2) Use its features as input to the model.
      - If there is no such row (not enough history, or outside data range),
        we fall back to the station+weekday mean baseline.

    Parameters
    ----------
    station_id : (int or str)
        Station ID. Must match the type in model_df['start_station_id'].
    target_date_str : str
        Desired prediction date as 'YYYY-MM-DD'.
    model_df : pd.DataFrame
        Prepared station-day panel with features and y_next_day.
    feature_cols : list of str
        Names of feature columns used for training the model.
    model : fitted estimator
        Your trained gradient-boosted model (e.g., gbr).
    use_baseline : bool
        Whether to fall back to baseline_station_weekday_mean if the ML model
        cannot be used for this (station, date).

    Returns
    -------
    dict
        {
          'station_id': ...,
          'target_date': Timestamp,
          'ml_prediction': float or None,
          'baseline_prediction': float or None
        }
    """
    # Parse target date and compute T-1
    target_date = pd.to_datetime(target_date_str)
    t_minus_1 = target_date - pd.Timedelta(days=1)

    # --- 1. Try to use the ML model (Gradient Boosted Trees) ---
    mask = (
        (model_df['start_station_id'] == station_id) &
        (model_df['start_date'] == t_minus_1)
    )
    row_today = model_df.loc[mask]

    ml_prediction = None
    if not row_today.empty:
        # Use this station-day's features to predict the next day
        X_new = row_today[feature_cols].values
        ml_prediction = float(model.predict(X_new)[0])

    # --- 2. Compute baseline prediction (station + weekday mean) ---
    baseline_prediction = None
    if use_baseline:
        tomorrow_dow = target_date.dayofweek  # 0=Mon,...,6=Sun
        baseline_input = pd.DataFrame({
            'start_station_id': [station_id],
            'day_of_week': [tomorrow_dow]
        })
        baseline_prediction = float(baseline_station_weekday_mean(baseline_input)[0])

    # --- 3. Print a friendly summary ---
    print("--------------------------------------------------")
    print(f"Station ID: {station_id}")
    print(f"Target date: {target_date.date()}")

    if ml_prediction is not None:
        print(f"ML model prediction (Gradient Boosted Trees): {ml_prediction:.2f} trips")
    else:
        print("ML model: cannot predict for this station/date (no row for T-1 in model_df).")

    if baseline_prediction is not None:
        print(f"Baseline (Station + Weekday Mean) prediction: {baseline_prediction:.2f} trips")
    else:
        print("Baseline: not available (unexpected).")

    if ml_prediction is None and baseline_prediction is not None:
        print("Using BASELINE prediction as fallback.")
    elif ml_prediction is None and baseline_prediction is None:
        print("No prediction available for this station/date.")
    else:
        print("You can compare ML vs baseline for this station/date.")

    print("--------------------------------------------------\n")

    return {
        'station_id': station_id,
        'target_date': target_date,
        'ml_prediction': ml_prediction,
        'baseline_prediction': baseline_prediction
    }


# -------------------------------------------------------------------
# Optional: "chatbot-style" loop in the notebook
# -------------------------------------------------------------------
# This lets you type a station_id and date and get predictions back.

import pandas.api.types as ptypes

def interactive_prediction_loop():
    """
    Simple CLI-style chatbot for predictions.
    Type station_id and date; get ML + baseline predictions.
    Type 'quit' to exit.
    """
    print("Interactive Capital Bikeshare predictor")
    print("Type 'quit' at any time to exit.\n")

    # Detect whether station_id is stored as int or string in your data
    station_is_int = ptypes.is_integer_dtype(model_df['start_station_id'])

    while True:
        user_station = input("Enter a station_id (or 'quit'): ").strip()
        if user_station.lower() in ["q", "quit", "exit"]:
            print("Goodbye!")
            break

        # Convert station_id to proper type
        if station_is_int:
            try:
                station_id = int(user_station)
            except ValueError:
                print("Please enter a numeric station_id.\n")
                continue
        else:
            station_id = user_station  # keep as string

        # Ask for target date
        user_date = input("Enter target date (YYYY-MM-DD) for prediction (or 'quit'): ").strip()
        if user_date.lower() in ["q", "quit", "exit"]:
            print("Goodbye!")
            break

        # Call the prediction helper
        try:
            _ = predict_trips_for_station_and_date(
                station_id=station_id,
                target_date_str=user_date,
                model_df=model_df,
                feature_cols=feature_cols,
                model=gbr,
                use_baseline=True
            )
        except Exception as e:
            print(f"Error while predicting: {e}\n")


# To use the interactive loop in your notebook, just run:
# interactive_prediction_loop()

predict_trips_for_station_and_date(
    station_id=31000,
    target_date_str="2025-10-31",
    model_df=model_df,
    feature_cols=feature_cols,
    model=gbr,
    use_baseline=True
)
```
